```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.implicitly_wait(3)
driver.get("https://www.livesport.com/kr/team/tottenham/UDg08Ohm/")
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
page=driver.page_source
```

```python
bs_obj = BeautifulSoup(page, "html.parser")
bs_obj
```

```python
win = bs_obj.find_all("div", {"class":"formIcon formIcon--w"})
win
```

```python
draw = bs_obj.find_all("div",{"class":"formIcon formIcon--d"})
draw
```

```python
lose = bs_obj.find_all("div",{"class":"formIcon formIcon--l"})
lose
```

```python
report ={'win':len(win), 'draw':len(draw), 'lose':len(lose)}
report
```

```python
max_n=max(report.values())
report.keys()
```

```python
for key in report:
    if(report[key]==max_n):
        print(key)
```
